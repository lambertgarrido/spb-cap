# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
def bootstrap_perms(concatenate_data, func, size_group_A, index_group_B):
    """Generate permutations. Calculate statistic of each group. Return difference"""
    perm = np.random.permutation(concatenate_data)
    group_A = perm[:size_group_A]
    group_B = perm[index_group_B:]
    return func(group_A) - func(group_B)
```

```python
# calculate mean and standard deviation for the excess rate of admissions
avg = np.mean(y)
std = np.std(y)
print('avg excess rate of admission = ', avg)
print('std excess rate of admission = ', std)

# calculate mean and std for number of discharges < 100
mask = np.array(x) < 100
y_lt100 = np.array(y)[mask]
avg_lt100 = np.mean(y_lt100)
std_lt100 = np.std (y_lt100)
print('avg excess rate of admission for # discharges < 100 = ', avg_lt100)
print('std excess rate of admission for # discharges < 100 = ', std_lt100)

mask = np.array(x) > 1000
y_gt1000 = np.array(y)[mask]
avg_gt1000 = np.mean(y_gt1000)
std_gt1000 = np.std (y_gt1000)
print('avg excess rate of admission for # discharges > 1000 = ', avg_gt1000)
print('std excess rate of admission for # discharges > 1000 = ', std_gt1000)

obs_diff = avg_lt100 - avg_gt1000
print('observed diff = ', obs_diff)

#assume that the excess rate of admission has the same distribution regardless of number of discharges
mask = (np.array(x) >= 100) & (np.array(x) <= 1000)
y_mid = np.array(y)[mask]

concatenate_data = np.concatenate((y_lt100, y_mid, y_gt1000))

count_lt100 = len(y_lt100)
count_gt1000= len(y_gt1000)
count_total = len(y)
index = count_total - count_gt1000

# take 100000 permutations of data
num_of_perms = 100000
bs_perms = np.empty(num_of_perms)
for i in range(num_of_perms):
    bs_perms[i] = bootstrap_perms(concatenate_data, np.mean, count_lt100, index)
    
p = np.sum(bs_perms >= obs_diff)/len(bs_perms)
print('p = ', p)
```

_It is assumed that the excess rate of admission distribution is the same for any hospital size (measured as number of discharges). Given this hypothesis, permutations of the data is performed 100,000 times and the difference of the excess rate of admission is calculated between the 2 groups. The first group consists of hospitals whose number of discharges is over 1000 and the second group consists of hospitals whose number of discharges is under 100. The observed difference between these 2 groups for the excess rate of admission is 0.0435. The probability of observing such a difference or bigger is less than 1 part in 100,000. Therefore the observed difference is statistically significant and our hypothesis is false._
